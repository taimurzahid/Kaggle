```python
def one_hot_current_employees(col_name):
    one_hot = pd.get_dummies(current_employees[str(col_name)])
    for name in one_hot.columns.values:
        one_hot.rename(columns={str(name): str(name) + ' ' + str(col_name)}, inplace=True)
    return one_hot
```

```python
def one_hot_former_employees(col_name):
    one_hot = pd.get_dummies(former_employees[str(col_name)])
    for name in one_hot.columns.values:
        one_hot.rename(columns={str(name): str(name) + ' ' + str(col_name)}, inplace=True)
    return one_hot
```
